```python
# 1. How many passengers are in our passenger list?   ANSWER: 891

# Get the number of records by pulling the length of the passenger id column:

total_passengers = len(data['PassengerId'])
print total_passengers
```

```python
# 2. What is the overall survival rate?   ANSWER: 38.4%

# number of people who survived / total number of passengers

# survivors:
survivors = data['Survived'].sum()
# survival rate:
survival_rate = (survivors / float(total_passengers))

print survival_rate
```
